```python
meta = MetaData()
```

```python
gpu = Table(
    "GPU",
    meta,
    Column('id', Integer, primary_key=True),
    Column('name', String, nullable=False),
    Column('vram', Integer, nullable=False),
    Column('cuda_core', Integer),
    Column('price', Integer, nullable=False)
)
meta.create_all(engine)
```

### ORM

```python
from sqlalchemy.orm import declarative_base, sessionmaker, relationship
from sqlalchemy import Float, Boolean
```

```python
eng = create_engine('sqlite:///orm.db', echo=True)
Base = declarative_base()
```

```python
class Architecture(Base):
    __tablename__ = 'architecture'

    id = Column(Integer, primary_key=True)
    decoder_blocks = Column(Integer, nullable=False)
    attn_type = Column(String)
    is_MOE = Column(Boolean)

    llms = relationship("LargeLM", back_populates="arch")

    def __repr__(self):
        return (
            f"Architecture("
            f"id={self.id}, "
            f"decoder_blocks={self.decoder_blocks}, "
            f"attn_type='{self.attn_type}', "
            f"is_MOE={self.is_MOE}"
            f")"
        )

class LargeLM(Base):
    __tablename__ = 'llm'

    id = Column(Integer, primary_key=True)
    name = Column(String, nullable=False)
    param_count_in_B = Column(Float, nullable=False)
    arch_id = Column(Integer, ForeignKey('architecture.id'))

    arch = relationship("Architecture", back_populates="llms")

    def __repr__(self):
        return (
            f"LargeLM("
            f"id={self.id}, "
            f"name='{self.name}', "
            f"param_count_in_B={self.param_count_in_B}, "
            f"arch_id={self.arch_id}"
            f")"
        )
```

```python
Base.metadata.create_all(eng)
```

```python
Session = sessionmaker(bind=eng)
session = Session()
```

```python
architectures = [
    Architecture(id=1, decoder_blocks=36, attn_type="GQA", is_MOE=False),
    Architecture(id=2, decoder_blocks=61, attn_type="MLA", is_MOE=True),
    Architecture(id=3, decoder_blocks=48, attn_type="GQA", is_MOE=False),
    Architecture(id=4, decoder_blocks=80, attn_type="MLA", is_MOE=True),
    Architecture(id=5, decoder_blocks=62, attn_type="GQA+SWA", is_MOE=False),
]
session.add_all(architectures)
session.commit()
```

```python
llms = [
    LargeLM(id=1, name="Qwen3-32B", param_count_in_B=32, arch_id=1),
    LargeLM(id=2, name="DeepSeek-V3", param_count_in_B=671, arch_id=2),
    LargeLM(id=3, name="Qwen3-14B", param_count_in_B=14, arch_id=1),
    LargeLM(id=4, name="Llama-4-Scout", param_count_in_B=109, arch_id=3),
    LargeLM(id=5, name="Kimi-K2", param_count_in_B=1000, arch_id=4),
    LargeLM(id=6, name="Gemma-3-27B", param_count_in_B=27, arch_id=5),
    LargeLM(id=7, name="Qwen3-8B", param_count_in_B=8, arch_id=1),
    LargeLM(id=8, name="DeepSeek-R1", param_count_in_B=671, arch_id=2),
    LargeLM(id=9, name="Gemma-3-12B", param_count_in_B=12, arch_id=5),
    LargeLM(id=10, name="Llama-4-Maverick", param_count_in_B=400, arch_id=3),
]
session.add_all(llms)
session.commit()
```
